```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df = pd.DataFrame(data=[[10,10],[20,20],[30,25]], columns=['time', 'acc'])
df1 = pd.DataFrame(data=[[10, 20], [20, 40], [40, 45]], columns=['time', 'acc'])
df_list = [['label1', df], ['label2', df1]]
```

```python
simplenet_list = []
resnet_list = []
vgg_list = []
efficientnet_list = []
for i in os.listdir('../src/results'):
    print(i)
    val = [i.split('.')[0], pd.read_csv('../src/results/'+i)]
    if 'simplenet' in i:
        simplenet_list.append(val)
    elif 'resnet18' in i:
        resnet_list.append(val)
    elif 'vgg' in i:
        vgg_list.append(val)
    elif 'efficientnet' in i:
        efficientnet_list.append(val)
```

```python
from scipy.interpolate import make_interp_spline, BSpline
for label, df in simplenet_list:
    T = df['time']
    power = df['acc']
    xnew = np.linspace(T.min(), T.max(), 300)  

    spl = make_interp_spline(T, power, k=3)  # type: BSpline
    power_smooth = spl(xnew)

    plt.plot(xnew,power_smooth, label=label)
#     plt.plot(, , label=label)
plt.title("Accuracy vs Time taken")
plt.xlabel("Time (sec)")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
